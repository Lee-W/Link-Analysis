```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.DiGraph()
```

```python
for line in open('dataset/graph_1.txt'):
    nodes = line.rstrip('\n').split(",")
    print(nodes)
    G.add_edge(nodes[0],nodes[1])
```
